Создание распределений

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.distributions import ECDF
norm_sample = np.random.normal(size=1000)
uniform_sample = np.random.uniform(size=1000)
expon_sample = np.random.exponential(size=1000)
bern_sample = stats.bernoulli.rvs(0.3, size=1000)
```

Описательные статистики нормального распределения

```python
norm_max = np.max(norm_sample)
print("Максимум нормального распределения - ", norm_max)

norm_min = np.min(norm_sample)
print("Минимум нормального распределения - ", norm_min)

norm_mean = np.mean(norm_sample)
print("Математическое ожидание нормального распределения - ", norm_mean)

norm_var = np.var(norm_sample)
print("Дисперсия нормального распределения - ", norm_var)

norm_quantil_75 = np.quantile(norm_sample, 0.75)
print("Квантиль 75% нормального распределения - ", norm_quantil_75)

norm_quantil_25 = np.quantile(norm_sample, 0.25)
print("Квантиль 75% нормального распределения - ", norm_quantil_25)

norm_quantil_75 = np.quantile(norm_sample, 0.75)
print("Квантиль 75% нормального распределения - ", norm_quantil_75)

norm_inq = norm_quantil_75 - norm_quantil_25
print("Межквантильный размах нормального распределения - ", norm_inq)

norm_median = np.median(norm_sample)
print("Медиана нормального распределения - ", norm_median)

norm_skew = stats.skew(norm_sample)
print("Коэффициент ассиметрии нормального распределения - ", norm_skew)

norm_kurtosis = stats.kurtosis(norm_sample)
print("Коэффициент эксцесса нормального распределения - ", norm_kurtosis)
```

Описательные статистики равномерного распределения

```python
uniform_max = np.max(uniform_sample)
print("Максимум равномерного распределения - ", uniform_max)

uniform_min = np.min(uniform_sample)
print("Минимум равномерного распределения - ", uniform_min)

uniform_mean = np.mean(uniform_sample)
print("Математическое ожидание равномерного распределения - ", uniform_mean)

uniform_var = np.var(uniform_sample)
print("Дисперсия равномерного распределения - ", uniform_var)

uniform_quantil_75 = np.quantile(uniform_sample, 0.75)
print("Квантиль 75% равномерного распределения - ", uniform_quantil_75)

uniform_quantil_25 = np.quantile(uniform_sample, 0.25)
print("Квантиль 25% равномерного распределения - ", uniform_quantil_25)

uniform_inq = uniform_quantil_75 - uniform_quantil_25
print("Межквантильный размах равномерного распределения - ", uniform_inq)

uniform_median = np.median(uniform_sample)
print("Медиана равномерного распределения - ", uniform_median)

uniform_skew = stats.skew(uniform_sample)
print("Коэффициент ассиметрии равномерного распределения - ", uniform_skew)

uniform_kurtosis = stats.kurtosis(uniform_sample)
print("Коэффициент эксцесса равномерного распределения - ", uniform_kurtosis)
```

Задание №2

```python
distributions = {}
sizes_of_distributions = [10, 50, 100, 1000]

for sizes in sizes_of_distributions:
    distributions[sizes] = np.random.normal(10, 2, size=(1000, sizes))
    average_values = np.mean(distributions[sizes], axis=1)  # Вычисляем средние значения для каждой выборки

    # Построение гистограммы средних значений
    plt.hist(average_values, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')

    # Настройка графика
    plt.title(f'Гистограмма средних значений выборок (размер {sizes})')
    plt.xlabel('Среднее значение')
    plt.ylabel('Плотность')

    plt.show()

    # Вычисляем выборочное среднее 
    sample_mean = np.mean(average_values)
    
    # Вычисляем стандартное отклонение
    sample_std_dev = np.std(average_values, ddof=1)
    
    # Вычисляем коэффициент вариации
    coefficient_of_variation = (sample_std_dev / sample_mean) * 100

    print(f"Размер выборки: {sizes}")
    print(f"Выборочное среднее: {sample_mean}")
    print(f"Выборочное стандартное отклонение: {sample_std_dev}")
    print(f"Коэффициент вариации: {coefficient_of_variation}%")
```

Выводы из данных: При увеличении объёма уменьшается отклонение данных от среднего значения, что делает их более пригодными для дальнейшей работы с этой выборкой. Уменьшение коэффициента вариации означает, что данные становятся более однородными и стабильными. Выборочное среднее изменяется незначительно.

Выборка стандартного распределения

```python
norm_stand = np.random.standard_normal(size=10000)
norm_distribution_ecdf = ECDF(norm_stand)
x = np.linspace(min(norm_stand), max(norm_stand))
y = norm_distribution_ecdf(x)
plt.title("Эмпирическая функция распределения")
plt.step(x, y)
plt.show()
```

```python
# Создаем массив значений x для вычисления эмпирической функции распределения
x_values = np.sort(norm_stand)

empirical_cdf = np.arange(1, 10000 + 1) / 10000

# Вычисляем значение статистики D_n для каждого размера выборки
D_n_values = np.abs(empirical_cdf - stats.norm.cdf(x_values))
D_n_max = np.max(D_n_values)

# Построение графика зависимости статистики D_n от n
plt.plot(range(1, 10000 + 1), D_n_values, color='r')
plt.xlabel('Размер выборки (n)')
plt.ylabel('Значение статистики D_n')
plt.title('Зависимость D_n от размера выборки')
plt.show()

# Вывод максимального значения D_n
print(f"Максимальное значение статистики D_n: {D_n_max}")
```

Выводы:
Dn уменьшается с увеличением выборки
